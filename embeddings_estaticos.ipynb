LSA

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD

# 1. Converter textos em matriz TF-IDF
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(documentos)

print("Formato original TF-IDF:")
print(X_tfidf.shape)   # (n_documentos, n_termos)
```

Q1) Observe o código abaixo e varie o número de componentes (vetores singulares).
Utilizando a similaridade coseno (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.cosine_similarity.html), encontre os documentos mais similares entre si. Os documentos resultantes se alteram?

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd

componentes = [2,3,5,7,9,11]

resultados = []

for n_componentes in componentes:
    # 2. Aplicar LSA
    lsa = TruncatedSVD(n_components=n_componentes, random_state=42)
    X_reduzido = lsa.fit_transform(X_tfidf)

    m_sim = cosine_similarity(X_reduzido)

    np.fill_diagonal(m_sim, -1)
    i, j = np.unravel_index(np.argmax(m_sim), m_sim.shape)

    resultados.append({
        "componentes" : n_componentes,
        "doc1" : i+1,
        "doc2" : j+1,
        "similaridade" : m_sim[i,j]
    })

df_q1 = pd.DataFrame(resultados)
df_q1
```

```python
# 3. Coordenadas dos documentos no espaço reduzido
print("\nRepresentação reduzida:")
for i, vetor in enumerate(X_reduzido[:10,:]):
    print(f"Doc {i+1}: {vetor}")
```

```python
# LSA para reduzir dimensionalidade de PALAVRAS

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
import pandas as pd

# 1. Matriz termo-documento
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documentos)

# termos (palavras)
termos = vectorizer.get_feature_names_out()
```

```python
# 2. Transpor para palavra x documento
# agora cada linha representa uma palavra
X_palavras = X.T

print("Formato original (palavras x documentos):")
print(X_palavras.shape)
```

Q2) Observe o código abaixo e varie o número de componentes (vetores singulares).
Utilizando a similaridade coseno (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.cosine_similarity.html), encontre as palavras mais similares entre si. Os documentos resultantes se alteram?

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd

componentes = [2,3,5,7,9,11]

resultados = []

for n_componentes in componentes:
    # 2. Aplicar LSA
    lsa = TruncatedSVD(n_components=n_componentes, random_state=42)
    X_palavras_reduzido = lsa.fit_transform(X_tfidf)

    m_sim = cosine_similarity(X_palavras_reduzido)

    np.fill_diagonal(m_sim, -1)
    i, j = np.unravel_index(np.argmax(m_sim), m_sim.shape)

    resultados.append({
        "componentes" : n_componentes,
        "doc1" : termos[i],
        "doc2" : termos[j],
        "similaridade" : m_sim[i,j]
    })

df_q2 = pd.DataFrame(resultados)
df_q2
```

```python
from sklearn.decomposition import TruncatedSVD

# 3. Aplicar LSA para palavras com 2 componentes para visualização
lsa_words = TruncatedSVD(n_components=2, random_state=42)
X_palavras_reduzido_2d = lsa_words.fit_transform(X_palavras)

# 4. Mostrar coordenadas reduzidas das palavras
df = pd.DataFrame(X_palavras_reduzido_2d, columns=["Dim1", "Dim2"])
df["Palavra"] = termos

print("\nVetores reduzidos das palavras:")
print(df[["Palavra", "Dim1", "Dim2"]].sort_values("Palavra"))
```

Q3) Altere o código abaixo para inicializar  avariável frases com os documentos anteriores tokenizados. Varie o tamanho da janela e do vetor no Word2Vec. Mostre como diferentes tamanhos podem alterar os resultados das palavras mais similares.

```python
from gensim.models import Word2Vec
import pandas as pd

frases = [doc.split() for doc in documentos]
frases

configuracoes = [
    {"vector_size" : 8, "window" : 2},
    {"vector_size" : 8, "window" : 4},
    {"vector_size" : 20, "window" : 2},
    {"vector_size" : 20, "window" : 4}
]

results = []

for config in configuracoes:
  modelo = Word2Vec(
      sentences=frases,
      vector_size=config["vector_size"],
      window=config["window"],
      min_count=1,
      epochs = 200,
      sg = 1,
      workers = 1,
      seed = 42
  )

  similars = modelo.wv.most_similar("gato", topn=5)

  results.append({
      "vector_size" : config["vector_size"],
      "window" : config["window"],
      "gato" : similars
      })

  df_q3 = pd.DataFrame(results)
  print(df_q3)
```

FastText

Q4) Varie o tamanho da janela e do vetor no FastText. Mostre como diferentes tamanhos podem alterar os resultados das palavras mais similares. Compare os resultados com aqueles obtidos pelo Word2Vec.

```python
from gensim.models import FastText
# Treinando modelo FastText
modelo = FastText(
    sentences=frases,
    vector_size=20,
    window=2,
    min_count=1,
    workers=1,
    sg=1              # Skip-Gram
)

# Vetor da palavra "gato"
print("Vetor de gato:")
print(modelo.wv["gato"])
```

```python
# Palavras mais similares
print("\nMais similares a gato:")
print(modelo.wv.most_similar("gato"))
```

Q5) Teste outras palavras fora do vocabulário e veja quais palavras o FastText retorna como mais similares.

```python

# Palavra não vista no treino
palavras_teste = ["gatinho", "cachorrinho", "computadorzinho", "casinha"]

for palavra in palavras_teste:
    print("\nPalavra testada:", palavra)
    print("Palavras mais similares:")

    similares = modelo.wv.most_similar(palavra, topn=5)

    for termo, score in similares:
        print(f"{termo} - {score:.4f}")
```

```python
import fasttext

# Carregar modelo pré-treinado em português
modelo = fasttext.load_model("/content/cc.pt.300.bin")

# Vetor da palavra
print("Vetor de gato:")
print(modelo.get_word_vector("gato")[:10])

# Palavra similar (via vizinhos mais próximos)
print("\nPalavras próximas de gato:")
print(modelo.get_nearest_neighbors("gato", k=5))

# Similaridade manual
import numpy as np

v1 = modelo.get_word_vector("gato")
v2 = modelo.get_word_vector("cachorro")

similaridade = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

print("\nSimilaridade gato x cachorro:")
print(similaridade)

# Palavra nunca vista (vantagem do FastText)
print("\nPalavra nova:")
print(modelo.get_word_vector("gatinhozinho")[:10])
```

```python
palavras_teste = ["gato", "cachorro", "computador", "gatinho"]

for palavra in palavras_teste:
    print("\nPalavra:", palavra)

    similares = modelo.get_nearest_neighbors(palavra, k=5)

    for score, termo in similares:
        print(f"{termo} ({score:.4f})")
```
